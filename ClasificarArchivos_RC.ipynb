```python
import os
import shutil
import pandas as pd
from PyPDF2 import PdfReader
```

```python
base_path = r"C:\Users\a.burgosg\OneDrive - Universidad de los andes\Revisiones de materias"
dest_path = r"C:\Users\a.burgosg\Documents\Historias Académicas"
excel_path = r"C:\Users\a.burgosg\Documents\Universidad de los Andes\Programas\Clasificación Archivos\Atributos_clas.xlsx"
df = pd.read_excel(excel_path) #Leer

carpeta_valores = []
for index, row in df.iterrows():
    carpeta = row['Carpeta']
    for col in df.columns[1:]:  # Omitir 1
        valor = row[col]
        if pd.notna(valor):  # Omitir NAN
            carpeta_valores.append((carpeta, str(valor)))
print("Valores extraídos del archivo de Excel para la clasificación:")
for carpeta, valor in carpeta_valores:
    print(f"Carpeta: {carpeta}, Valor: {valor}")
    
def extract_text_from_first_page(pdf_path):
    try:
        with open(pdf_path, "rb") as file:
            reader = PdfReader(file)
            if len(reader.pages) > 0:
                page = reader.pages[0]
                return page.extract_text()
    except Exception as e:
        print(f"Error al leer el archivo PDF: {e}")
    return ""


def clasificar_archivo_por_contenido(pdf_path, carpeta_valores, dest_path):
    contenido = extract_text_from_first_page(pdf_path)
    for carpeta, valor in carpeta_valores:
        if valor in contenido:
            nueva_carpeta = os.path.join(dest_path, carpeta)
            if not os.path.exists(nueva_carpeta):
                os.makedirs(nueva_carpeta)
            shutil.copy(pdf_path, nueva_carpeta)
            print(f"Archivo {os.path.basename(pdf_path)} copiado a {nueva_carpeta}")
            return True
    return False


def procesar_carpeta(base_path, carpeta_valores, dest_path):
    for root, dirs, files in os.walk(base_path):
        for file in files:
            file_path = os.path.join(root, file)
            if file_path.lower().endswith('.pdf'):
                if not clasificar_archivo_por_contenido(file_path, carpeta_valores, dest_path):
                    print(f"No se encontró coincidencia significativa para el archivo: {os.path.basename(file_path)}")

procesar_carpeta(base_path, carpeta_valores, dest_path)
```
